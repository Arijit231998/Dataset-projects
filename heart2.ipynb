Exploratory Data Analysis

```python
x=sb.countplot(x='RestingBP',data=df)
sb.set(rc={'figure.figsize':(80,70)})
for bars in x.containers:
    x.bar_label(bars,fontsize=100)
```

```python
x=sb.countplot(x='Cholesterol',data=df)
sb.set(rc={'figure.figsize':(100,80)})
for bars in x.containers:
    x.bar_label(bars,fontsize=60)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the figure size to a more reasonable size
plt.figure(figsize=(12, 8))  # Adjust to your preference

# Create the scatter plot
sns.scatterplot(x='Age', y='RestingBP', data=df)

# Add a title
plt.title('Scatter Plot of Age vs RestingBP')

# Show the plot
plt.show()
```
