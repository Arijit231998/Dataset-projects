```python
import pandas as pd
import numpy as np
import matplotlib as plt 
import seaborn as sb
```

```python
df=pd.read_csv(r"fetal_health.csv",encoding='unicode_escape')
```

```python
columns_to_convert = [
    'baseline value', 'accelerations', 'fetal_movement',
    'uterine_contractions', 'light_decelerations', 'severe_decelerations',
    'prolongued_decelerations', 'abnormal_short_term_variability',
    'mean_value_of_short_term_variability',
    'percentage_of_time_with_abnormal_long_term_variability',
    'mean_value_of_long_term_variability', 'histogram_width',
    'histogram_min', 'histogram_max', 'histogram_number_of_peaks',
    'histogram_number_of_zeroes', 'histogram_mode', 'histogram_mean',
    'histogram_median', 'histogram_variance', 'histogram_tendency',
    'fetal_health'
]

# Convert the specified columns to integer type
df[columns_to_convert] = df[columns_to_convert].astype(int)
```

```python
df.info()
```

```python
df.shape
```

```python
pd.isnull(df)
```

```python
null_counts=pd.isnull(df).sum()
print(null_counts)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df.describe()
```

```python
x=sb.countplot(x=  'baseline value',data=df)
sb.set(rc={'figure.figsize':(96,10)})
for bars in x.containers:
    x.bar_label(bars,fontsize=100)
```

```python
x=sb.countplot(x=  'accelerations',data=df)
sb.set(rc={'figure.figsize':(80,3)})
for bars in x.containers:
    x.bar_label(bars,fontsize=100)
```

```python
import matplotlib.pyplot as plt
x=sb.barplot(x='baseline value',y='abnormal_short_term_variability',data=df)
sb.set(rc={'figure.figsize':(90,50)})
plt.show()
```
